### Party Performance - Total Seats and Vote Share

```python
seats_won = df['Leading Party'].value_counts().reset_index()
seats_won.columns = ['Party', 'Seats Won']

total_votes = df['Margin'].sum()

partywise_votes = df.groupby('Leading Party')['Margin'].sum().reset_index()

partywise_votes['Vote Share (%)'] = (partywise_votes['Margin'] / total_votes) * 100

partywise_performance = pd.merge(seats_won, partywise_votes, left_on='Party', right_on='Leading Party')
partywise_performance = partywise_performance[['Party', 'Seats Won', 'Vote Share (%)']]

partywise_performance
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 12))
plt.barh(partywise_performance['Party'], partywise_performance['Seats Won'], color='red')
plt.ylabel('Party')

plt.xlabel('Seats Won')
plt.title('Total Seats Won by Each Party')

plt.gca().invert_yaxis()
plt.show()
```

### Seat Distribution by Party within State

### Statewise Leading Party and Seats Won

```python
statewise_seats = df.groupby(['State and Union Territory', 'Leading Party']).size().unstack(fill_value=0)

leading_party_per_state = statewise_seats.idxmax(axis=1)
seats_won_per_state = statewise_seats.max(axis=1)

result = pd.DataFrame({'State and Union Territory': leading_party_per_state.index,
                       'Leading Party': leading_party_per_state.values,
                       'Seats Won': seats_won_per_state.values})
result
```

```python
result.plot(x="State and Union Territory", y="Seats Won", kind="bar")
plt.xlabel("State and Union Territory")
plt.ylabel("Seats Won")
plt.title("Seats Won by Leading Party in Each State")
plt.show()
```

### BJP Performance in Every State

```python
bjp_perf = df[df['Leading Party'] == 'Bharatiya Janata Party']

bjp_s = bjp_perf['State and Union Territory'].value_counts()
all_states = df['State and Union Territory'].unique()

all_states_df = pd.DataFrame(all_states, columns=['State and Union Territory'])

display = all_states_df.merge(bjp_s, left_on='State and Union Territory', right_index=True, how='left')
display.columns = ['State and Union Territory', 'Seats Won']
display['Seats Won'] = display['Seats Won'].fillna(0).astype(int)

x=display.sort_values(by='Seats Won', ascending=False).reset_index()

x
```

```python

plt.figure(figsize=(12, 8))
plt.bar(x['State and Union Territory'], x['Seats Won'], color='orange')
plt.title('BJP Performance in All States')
plt.xlabel('State and Union Territory')
plt.ylabel('Seats Won')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
